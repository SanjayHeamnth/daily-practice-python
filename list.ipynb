##### list comprahention
##### basic syntax [expression for item in  ittrable ]
##### with conditional logic [expresion for iten in ittrable if condititon]
##### nested list comprahention [expresion for item1 in ittrable 1 for itme2 in ittrable2]

```python
# basic comprahention
square = [x**2 for x in range(10)]
print(square)
```

```python
# normal code
list = []
for x in range(10):
    if x%2==0:
        list.append(x**2)
print(list)
```

```python
# lsit comprahention with condition
square_numebr = [x**2 for x in range(10) if x%2==0]
print(square_numebr)
```

```python
# nested list Comprahention
list1 = [1,2,3,4]
list2 = ['a','b','c','d']
pair = [[i,j] for i in list1 for j in list2]
print(pair)
```
