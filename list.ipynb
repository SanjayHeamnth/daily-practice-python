```python
 # creating infinite list
from itertools import count

inf_list = count(start=0, step=1)  # Infinite iterator (0, 1, 2, ...)

# Usage:
print(next(inf_list))  # 0
print(next(inf_list))  # 1
# ... runs forever
```

```python
# othrer way of wrtiting infinite list
def infinite_list():
    num = 0
    while True:  # Infinite loop (but lazy evaluation)
        yield num
        num += 1

# Usage:
inf_gen = infinite_list()  # Doesn't store all elements in memory
print(next(inf_gen))  # 0
print(next(inf_gen))  # 1
print(next(inf_gen))  # 2
# ... continues infinitely
```

```python
# saparating the numerical and catagoricals,  unpacking other lists tuple  and dictionary in th e bog list
def separate_nested_data(data):
    numerical = []
    categorical = []

    def _flatten(item):
        if isinstance(item, (int, float)) and not isinstance(item, bool):
            numerical.append(item)
        elif isinstance(item, (str, bool, type(None))):  # Basic categorical types
            categorical.append(item)
        elif isinstance(item, (list, tuple, set)):  # Unpack iterables
            for sub_item in item:
                _flatten(sub_item)
        elif isinstance(item, dict):  # Unpack dicts (check keys/values)
            for key, value in item.items():
                _flatten(key)
                _flatten(value)
        else:  # Other types (custom objects, etc.)
            categorical.append(item)

    _flatten(data)
    return numerical, categorical
```

```python
separate_nested_data(mixed_list)
```

```python
def separate_data2(data):
    numerical = []
    categorical = []
    
    def process(item):
        if isinstance(item, (int, float)) and not isinstance(item, bool):
            numerical.append(item)
        elif isinstance(item, (str, bool, type(None))):
            categorical.append(item)
        elif isinstance(item, dict):
            process(list(item.keys()) + list(item.values()))
        elif isinstance(item, (list, tuple, set)):
            for x in item:
                process(x)
        else:
            categorical.append(item)
    
    process(data)
    return numerical, categorical
```
