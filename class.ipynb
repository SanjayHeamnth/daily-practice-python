The __str__() method controls what should be returned when the class object is represented as a string.

```python
# The string representation of an object WITHOUT the __str__() method:
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

p11 = Person("sanjay",33)

print(p11.name)
```

```python
# The string representation of an object WITH the __str__() method:
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"{self.name},({self.age})"
        

p11 = Person("sanjay",33)

print(p11)
```

## difference btw __init__ and __str__

| Feature      | `__init__`                           | `__str__`                                                    |
| ------------ | ------------------------------------ | ------------------------------------------------------------ |
| When called  | Automatically when object is created | When object is printed or converted to string                |
| Purpose      | Initialize the objectâ€™s attributes   | Provide a human-readable string representation of the object |
| Return value | None                                 | Must return a string                                         |
| Role         | Sets up the object                   | Displays the object in a readable form                       |

What is __init__?
The __init__ method is a special method in Python classes, often referred to as the constructor. Its primary purpose is to initialize the attributes of an object when it's created. When you create an instance of a class, Python automatically calls __init__ to set up the object's initial state.

How to Define __init__
You define __init__ just like any other method in a class, with a few key rules:

Method Signature: It must be named __init__. The name is fixed, with two underscores at the beginning and two at the end.

The self Parameter: The first parameter must always be self. This parameter is a reference to the instance of the object that's being created. It allows you to access and set the object's attributes. You don't pass self explicitly when you create the object; Python handles this automatically.

Additional Parameters: After self, you can include any other parameters needed to initialize the object's attributes. These parameters are used to pass in the data that defines the object.

Attribute Assignment: Inside the method, you use the dot notation with self to assign the passed parameters to the object's attributes (e.g., self.brand = brand).

The __str__ method is a special method in a Python class that provides a human-readable string representation of an object. Its primary purpose is to make an object's information easy to read and understand, especially for developers and users.

When you use the print() function on an object or call the built-in str() function, Python automatically looks for and calls the object's __str__ method to get the string it should display.

Defining and Using __str__
The __str__ method must:

Accept one parameter: The first and only parameter must be self, which refers to the instance of the class.

Return a string: It must return a string. It cannot return any other data type or use print() internally to display information.

```python
class cc:
    def __init__(self,r):
        self.r = r

    def __str__(self):
        return f"curcumfarance: {2*3.14*self.r}" #You need to use curly braces {} in an f-string to perform an action, such as a calculation or accessing a variable.


ccc = cc(3)
print(ccc)
```
