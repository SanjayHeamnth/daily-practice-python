# Tuples

Tuples are ordered collection of items that are immutable , Similar to lists, but there immutable nature make  them different

```python
# creating a empty_tule
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
# other way fo creating tuples
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers
```

```python
mixed_tuple = (1,"hey",3.14,True)
```

```python
# Accesing the tuple elements
numbers
```

```python
# indexing and slicing are similar to list
numbers[0]
```

```python
# Tuple oparation
numbers
```

```python
concantenation = numbers + mixed_tuple
print(concantenation)
```

```python
mixed_tuple*3
```

```python
# immutable Nature of tuple
# tuples are imutable in nature
numbers[2] = 'rocky' # tuples are immutable
```

```python
# tuple methods
print(numbers.count(1))
```

```python
print(numbers.index(3))
```

```python
# packing and unpacking tuples
packing_tuple1 = 1,'hey',3.14  # it will automaticaly gets  packed  to tupple
print(packing_tuple1)
```

```python
# unpacking tuple
a,b,c, = packing_tuple1 
print(a)
print(b)
print(c)
```

```python
# unpacking with *
numbers = (1,2,3,4,5,6) # * make ti such a way  like aft the 1st element till last element will be consedred as middle
first,*middle,last =  numbers
print(first)
print(middle)
print(last)
```

```python
# nested_Tuples
nested_Tuple = ((1,2,3), ('a','b','c'), (True,False))

# accessing the element in tuple
print(nested_Tuple[0])
```

```python
print(nested_Tuple[0][2])
```

```python
# iitarating over nested tuples
for sub_tuple in nested_Tuple:
    for item in sub_tuple:
        print(item,end = '')
    print()
```
