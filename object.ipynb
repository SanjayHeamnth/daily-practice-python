What is an object in Python OOP?

In OOP (Object-Oriented Programming), an object is simply a real-world entity represented in code.

Think of it like this:

Class = Blueprint (recipe / design).

Object = Actual thing created from that blueprint.

For example:

Class: Car (blueprint says cars have color, model, speed).

Object: my_car = Car("Red", "Tesla", 200) (your actual car with details).

```python
# Defining a class (blueprint)
class Car:
    def __init__(self, color, model, speed):
        self.color = color
        self.model = model
        self.speed = speed
    
    def show_details(self):
        print(f"My car is a {self.color} {self.model} running at {self.speed} km/h")

# Creating objects (real cars from blueprint)
car1 = Car("Red", "Tesla", 200)
car2 = Car("Blue", "BMW", 180)

# Using the objects
car1.show_details()
car2.show_details()
```
