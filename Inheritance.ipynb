## Python Inheritance
Inheritance allows us to define a class that inherits all the methods and properties from another class.

Parent class is the class being inherited from, also called base class.

Child class is the class that inherits from another class, also called derived class.



Example

Get your own Python Server


Create a class named Person, with firstname and lastname properties, and a printname method:

```python
class person:
    def __init__(self, fname, lname):
        self.fname = fname
        self.lname = lname

    def printname(self):
        print(self.fname, self.lname)


x = person("Sanjay","hemanth")
x.printname()
```

Create a Child Class
To create a class that inherits the functionality from another class, 

send the parent class as a parameter when creating the child class:

```python
class student(person):
    pass
```

```python
x = student("Mike", "Olsen")
x.printname()
```

```python
# single inheritance
class Dog:   # Best practice: Class names start with Capital letters
    def __init__(self, name):
        self.name = name

    def displayname(self):
        print(f"Dog's name: {self.name}")

class Labrador(Dog):  # single inheritance
    def sound(self):
        print("Lab woofs!")

du = Labrador("Karna")
du.displayname()   # Inherited from Dog class
du.sound()         # Method from Labrador class
```

```python
# multilevel inheritance
class GuideDog(Labrador):
    def guide(self):
        print(f"{self.name} Guides the way")
```

```python
dp = GuideDog("kali")
dp.guide()
dp.displayname()
dp.sound()
```

```python
# Multipal Inheritance
class friendly:
    def greet(self):
        print("friendly")

class GoldenRetriver(dog, friendly, GuideDog): # can add multipal different class
    def sound(self):
        print("GoldenRetriver Barks!")
```

```python
df = GoldenRetriver("Browny")
df.sound()
df.displayname
df.greet()
df.guide()
```

What is Hierarchical Inheritance?

In hierarchical inheritance, one parent class is inherited by multiple child classes.

Think of a family tree:

Parent = Animal

Children = Dog, Cat, Cow (all inherit features from Animal).

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

# Child class 1
class Dog(Animal):
    def sound(self):
        print(f"{self.name} says Woof!")

# Child class 2
class Cat(Animal):
    def sound(self):
        print(f"{self.name} says Meow!")

# Child class 3
class Cow(Animal):
    def sound(self):
        print(f"{self.name} says Moo!")

# Creating objects from different child classes
dog1 = Dog("Bruno")
cat1 = Cat("Kitty")
cow1 = Cow("Ganga")

# Using parent and child methods
dog1.eat()
dog1.sound()

cat1.eat()
cat1.sound()

cow1.eat()
cow1.sound()
```

What is Hybrid Inheritance?

Hybrid inheritance = a combination of two or more types of inheritance.

Example: Mixing single, multiple, and hierarchical inheritance in one structure.

Think of it like this:

Parent class → Person

Child classes → Student, Employee (both inherit from Person → hierarchical inheritance)

Another class → Intern (inherits from both Student and Employee → multiple inheritance)

So the whole structure = Hybrid.

```python
# Base class
class Person:
    def __init__(self, name):
        self.name = name

    def display(self):
        print(f"Name: {self.name}")

# Hierarchical Inheritance
class Student(Person):
    def study(self):
        print(f"{self.name} is studying.")

class Employee(Person):
    def work(self):
        print(f"{self.name} is working.")

# Hybrid Inheritance (Multiple + Hierarchical)
class Intern(Student, Employee):
    def intern_role(self):
        print(f"{self.name} is an intern (studying + working).")

# Creating objects
intern1 = Intern("Karna")

intern1.display()      # From Person
intern1.study()        # From Student
intern1.work()         # From Employee
intern1.intern_role()  # From Intern
```
