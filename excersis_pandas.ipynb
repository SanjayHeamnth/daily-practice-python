### Create Pandas Dataframe from 2D List using pd.DataFrame()

```python
import pandas as pd
lis =[ ['sanjay',25],['abhi',21],
      ['rakesh',89],['eshan',78]]

edt = pd.DataFrame(lis,columns = ["tag",'number'])
edt
```

### Create Pandas Dataframe from 2D List using pd.DataFrame.from_records()


```python
columns = ['Name','age']

dd = pd.DataFrame.from_records(lis,columns = columns)
dd
```

### Create Pandas Dataframe from 2D List using pd.DataFrame.from_dict()

```python
lis =[ ['sanjay',25],['abhi',21],
      ['rakesh',89],['eshan',78]]

dic = pd.DataFrame.from_dict(dict(zip(columns,zip(lis))))
dic
```

Using Pandas Dataframe with the index parameter


```python
import pandas as pd 

# initialise data of lists. 
data = {'Area':['Array', 'Stack', 'Queue'],
        'Student_1':[20, 21, 19], 'Student_2':[15, 20, 14]} 

# Create DataFrame 
df = pd.DataFrame(data, index =['Cat_1', 'Cat_2', 'Cat_3']) 

# Print the output. 
print(df)
```

#### Handling Missing Values


```python
import numpy as np

# Creating a DataFrame with missing values from a list of lists
data = [['Geek1', 28, 'Engineer'],
        ['Geek2', None, 'Data Scientist'],
        ['Geek3', 32, None]]

colunns = ['name','age','accupation']
df = pd.DataFrame(data,columns = colunns)
df = df.replace({None:np.nan})
df
```

```python
import pandas as pd
import numpy as np

data = {
    'Employee': ['E1','E2','E3','E4','E5','E6','E7','E8'],
    'Sales': [120, 80, 200, 150, 90, 300, 60, 400],
    'Experience': [2, 5, 3, 10, 1, 7, 2, 12],
    'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'C']
}
df = pd.DataFrame(data)
print(df)
```

```python
results = []
for  row in df.itertuples(index = False):
    if row.Category == "A":
        bonus = row.Sales*row.Experience
    elif row.Category =="B":
        bonus =  row.Sales*row.Experience
    elif row.Category == "C":
        bonus = (row.Sales ** 2) / (row.Experience + 1)
    results.append(bonus)

df['bonus']= results
df
```

```python
import pandas as pd

record = {

 'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
 'Age': [21, 19, 20, 18, 17, 21],
 'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
 'Percentage': [88, 92, 95, 70, 65, 78] }

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

print("Given Dataframe :\n", dataframe) 

# selecting rows based on condition
rslt_df = dataframe[dataframe['Percentage']>80]

print('\nResult dataframe :\n', rslt_df)
```

```python
import pandas as pd 
  
# Create the dataframe 
df = pd.DataFrame({'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/11'], 
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'], 
                    'Cost':[10000, 5000, 15000, 2000]}) 

# Create an empty list 
Row_list =[] 
  
# Iterate over each row 
for i in range((df.shape[0])): 
  
    # Using iloc to access the values of  
    # the current row denoted by "i" 
    Row_list.append(list(df.iloc[i, :])) 
  
# Print the first 3 elements 
print(Row_list[:3])
```
