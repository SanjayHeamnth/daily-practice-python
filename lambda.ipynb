## Lambda Function

Lambda fucntions are small anonymous fucntion defined using lamda keywrd
they can have any number of arguments but only one expresion
they are used  or short oparation or arguments for higer order function


# syntax
lambda arguments:expresion


```python
addition = lambda a,b:a+b
type(addition)
print(addition(2,3))
```

```python
even = lambda num:num%2==0 # can have multipal agruments
print(even(2))
```

```python
# map() applies the function to all elements to a list 
numbers = [1,2,3,4,5,6]
list(map(lambda x:x**2,numbers))
```
