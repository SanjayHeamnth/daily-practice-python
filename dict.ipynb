# Dictionary

Dict are unordered collection of items. they store data in key-value pairs where keys are unique and immutable(eg:str,numbers or tuples),which values can be of any type

```python
# creating a empty dictionary 
empt_dict = {}
print(type(empt_dict))
```

```python
# creating a dict
student = {'name':'Sanjay', 'age':27, 'grade':3}
print(student)
```

```python
# slinge key is used
stu =  {'name':'Sanjay', 'age':27, 'name':3}# only with unique key the values are stored
stu
```

```python
# accessing the dictionary
student = {'name':'Sanjay', 'age':27, 'grade':'A'}
student
```

```python
print(student['grade'])
print(student['age'])
```

```python
# accessing through get method
print(student.get('name'))
print(student.get('lat_name'))# if the key is not present it will give none
print(student.get('last_name','note_avalible'))#  here i can get it as not avalible
```

```python
# Modifying the Dictionary Elements
# Dictionary are mutable we can add update and delete elements but key havs to be unique
student['age'] = 25 # updating the dictionary
print(student)
student['addres'] = 'india'  # adding a element to dictionary
print(student)
```

```python
# deleting a key 
del student['grade']
student
```

```python
# Dictionary methods
keys = student.keys()# get all the keys
print(keys)
values = student.values() # get all values
print(values)
```

```python

items = student.items() # will get all value paries
items
```

```python
# shallow_copy
student_copy = student
print(student_copy)
print(student)
```

```python
student['name'] = 'sanjay.H' # changes made will reflect in the copy as well to avoid this  we will use shallow copy
print(student_copy)
print(student)
```

```python
student_copy1 = student.copy() # shallow copy
print(student_copy1)
print(student)
```

```python
student['name'] = 'sanjay_' # changes are not updated 
print(student_copy1)
print(student)
```

```python
# ittrating the dictionares
for x in student.keys():
    print(x)
```

```python
for x in student.values():
    print(x)
```

```python
for x,y in student.items():
    print(f'{x}:{y}')
```

```python
# Nestesd dictionary
N_student = {
    "student1":{"name":"Rakesh", 'age':40},
    "student2":{"name":"Abhi", 'age':50}
}
N_student
```

```python
# accesing the neted dictionary
print(N_student['student2']['name'])
```

```python
# ittratitiong the nested dictionary
for student_id,student_info in student.items():
    print(f"{Student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")
```

```python
# Dictionary Comprahension
squares = {x:x**2 for x in range(5)}
squares
```

```python
# Condition dictionary comprahension
evens = {x:x**2 for x in range(10) if x%2==0} # gives squares of only evens
evens
```

```python
# use a Dictionary to count the frequency of number  in list
numbers = [1,2,2,2,3,4,3,4]
frequency = {}

for numbers in numbers:
    if numbers in frequency:
        frequency[numbers]+=1
    else:
        frequency[numbers] = 1
print(frequency)
```

```python
# Merg 2 Dictionaries inot one
dict1 = {'a':1, 'b':2}
dict2 = {'b':3, 'c':4}

merg_dict = {**dict1,**dict2}
print(merg_dict)
```
