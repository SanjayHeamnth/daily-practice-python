## Function Examples

#### Example 1 Temparature conversion

```python
def temp_converter(temp,unit):
    if unit=="c":
        return temp*9/+32
    elif unit=='F':
        return (temp-32)*5/9
    else:
        return None

```

```python
temp_converter(32,'f')
```

#### Example 2 Password Strength checker

```python
def is_strong_password(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any (char in "!@#$_%^&*()+" for char in password):
        return False
    return True
```

```python
is_strong_password("Weakpsw")
```

### Example_3 calculate the total cost of the items in a shoping cart


```python
def calculator(cart):
    total_cost = 0
    for item in cart:
        total_cost+=item['price']*item['quantity']
    return total_cost

cart = [
    {'name':'apple', 'price':0.4, 'quantity':3},
    {'name':'banna', 'price':0.5, 'quantity':3},
    {'name':'carrot', 'price':0.2, 'quantity':3}
]

#function calling
total = calculator(cart)
total
```

### Example_4 Check if a string si palindrom


```python
def palindrom(s):
    s = s.lower().replace(" ","")
    return s==s[::-1]

print(palindrom('abba'))
print(palindrom('deer'))
```

### Example_5 : Calculate the Factorial of a number using recursion


```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(3))
```
